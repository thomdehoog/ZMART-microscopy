# Set stage limits

Set the microscope's **physical stage envelope**: the hard XY / Z ranges the
driver refuses to move outside, in micrometers. These are hardware safety
limits for this scope, not a per-experiment area (that is the workflow's
concern).

## Enter the envelope and publish

Fill in each axis `[min, max]` in micrometers, then run the cell. It validates
the envelope against the canonical schema and writes it to the committed
`defaults.json` for this scope. The driver reads that file by default; the
target-acquisition workflow maintains a separate runtime `current.json` from
boundary markers or scan-field geometry. Commit `defaults.json` after running.

```python
import _bootstrap  # noqa: F401  — adds the driver + microscopes roots to sys.path
from navigator_expert.stage import config as stage_config

# Physical stage envelope for THIS microscope, in micrometers: the hard
# XY / Z ranges the driver refuses to move outside. These are hardware
# safety limits for the scope, not a per-experiment area (that is the
# workflow's concern). Edit each axis [min, max], then run the cell.
stage_um = {
    "x": [1000, 130000],
    "y": [1000, 100000],
    "z_galvo": [-200, 200],
    "z_wide": [0, 25000],
}

# Publish as the committed physical envelope (defaults.json). The loader
# reads this file by default; the workflow's per-run current.json is a
# separate, runtime-only concern. Commit defaults.json after running so
# the scope's limits travel with the repository.
path = stage_config.write_limits(
    stage_um,
    source=stage_config.LIMITS_SOURCE_DEFAULTS,
    path=stage_config.defaults_path(),
)
print(f"Wrote {path}")
print("Remember to `git add` and commit this file.")
```
