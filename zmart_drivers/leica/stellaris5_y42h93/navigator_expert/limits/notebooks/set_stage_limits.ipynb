# Set stage limits

Set the microscope's **physical stage envelope**: the hard XY / Z ranges the
driver refuses to move outside, in micrometers. These are hardware safety
limits for this scope, not a per-experiment area (that is the workflow's
concern).

## Step 1: Enter the envelope and publish

Fill in each axis `[min, max]` in micrometers, then run the cell. It
validates the envelope and publishes it as a dated snapshot under
`C:\\ProgramData\\zmart-microscopy\\...`, archiving this executed notebook.
The driver reads the newest snapshot.

```python
import _bootstrap
from datetime import datetime, timezone
from navigator_expert.calibration.core.common import make_session_paths
from navigator_expert.config.machine import MACHINE, format_snapshot_name
from navigator_expert.motion import stage_config

SESSIONS_ROOT = MACHINE.snapshot_root() / "limits"
make_session_paths(format_snapshot_name(datetime.now(timezone.utc)), "limits", SESSIONS_ROOT)

stage_um = {
    "x": [1000, 130000],
    "y": [1000, 100000],
    "z_galvo": [-200, 200],
    "z_wide": [0, 25000],
}

result = stage_config.adopt_limits(
    stage_um,
    notebook_paths=[_bootstrap.NOTEBOOKS_DIR / "set_stage_limits.ipynb"],
)
print(result)
```
