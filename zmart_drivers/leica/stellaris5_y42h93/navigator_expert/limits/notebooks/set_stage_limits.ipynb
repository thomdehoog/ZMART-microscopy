# Set stage limits

Set the microscope's **physical stage envelope**: the hard XY / Z ranges the
driver refuses to move outside, in micrometers. These are hardware safety
limits for this scope, not a per-experiment area (that is the workflow's
concern).

## Step 1: Enter the envelope and publish

Fill in each axis `[min, max]` in micrometers, then run the cell. It
validates the envelope and publishes it as a dated snapshot under
`C:\\ProgramData\\zmart-microscopy\\...`, archiving this executed notebook.
The driver reads the newest snapshot.

This writes the **single** `limits.json` — the function-keyed limits file
(`constraints` = the stage envelope stated by name + `functions` = the gate
policy). There is no separate `function_limits.json`, and no `backlash`
block: backlash is a plain motion utility with baked-in defaults (decision
§2b), not config. Both readers (the motion check and the commands gate)
read this one file.

Each snapshot dir holds exactly three files: `limits.json`,
`calibration.json`, `origin.json`. A limits adopt writes **only**
`limits.json`: it carries a *real* prior `calibration.json` forward if one
exists but never mints one from the bundled template (calibration stays a
loud read-time fallback until an explicit calibration adopt).


```python
import _bootstrap
from navigator_expert.motion import stage_config

stage_um = {
    "x": [1000, 130000],
    "y": [1000, 100000],
    "z_galvo": [-200, 200],
    "z_wide": [0, 25000],
}

result = stage_config.adopt_limits(
    stage_um,
    notebook_paths=[_bootstrap.NOTEBOOKS_DIR / "set_stage_limits.ipynb"],
)
print(result)
```
