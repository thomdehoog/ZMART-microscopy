Pick the Leica instrument, tell it where `acquire()` should save, and connect. Connecting also applies the machine's stage safety limits.

```python
instrument = next(i for i in zmart_controller.get_instruments() if i["vendor"] == "leica")
instrument["output_root"] = r"D:\zmart_output"  # <-- edit: where acquire() saves
zmart_controller.set_instrument(instrument)
```

## Move & acquire (moves the stage!)

`set_origin()` zeroes the frame at the current position; `set_xyz` then works in micrometres from there. Focus (`z`) is the sum of the two drives; pick one with `with_actuators`.

```python
zmart_controller.set_origin()
```

```python
zmart_controller.set_xyz(25, 25, 2, with_actuators={"z": "z-galvo"})
zmart_controller.get_xyz()
```

```python
zmart_controller.acquire(acquisition_type="demo-scan", position_label="1")
```

```python
zmart_controller.set_xyz(0, 0, 0, with_actuators={"z": "z-galvo"})  # back to the origin
```

```python
zmart_controller.disconnect()
```
