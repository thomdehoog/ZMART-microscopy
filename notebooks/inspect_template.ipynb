# Microscopy Template Inspection

Parses, enriches, and visualises a microscope scanning template.

Uses the manufacturer-agnostic `initialize_experiment()` entry point.
The `BACKEND` parameter selects the microscope vendor
(currently `"lasx"` for Leica LAS X).

**Enrichment modes:**
- `"auto"` — tries API first, then image files, then none
- `"api"` — requires a running microscope instance with the API enabled
- `"files"` — reads pixel sizes from OME-TIFF image headers on disk
- `"none"` — template structure only (tile sizes may be unknown)

```python
%matplotlib inline
import sys
from pathlib import Path

sys.path.insert(0, str((Path("..") / "lib").resolve()))

from microscope_inspect import initialize_experiment

# -- Parameters ---------------------------------------------------------------
BACKEND         = "lasx"        # microscope vendor backend
INPUT           = "auto"        # "auto" | path to template folder | None for picker
ENRICH          = "auto"        # "auto" | "api" | "files" | "none"
EXPERIMENT_ROOT = None          # experiment root for file-based enrichment
OUTPUT_PNG      = None          # save figure + JSON, e.g. "./out.png"
FIGSIZE         = (14, 10)      # figure size (width, height) in inches
VERBOSE         = 1             # 0 = data only | 1 = summary + figure | 2 = full report

# -- Run ---------------------------------------------------------------------
data = initialize_experiment(
    BACKEND,
    INPUT,
    enrich=ENRICH,
    experiment_root=EXPERIMENT_ROOT,
    output_path=OUTPUT_PNG,
    figsize=FIGSIZE,
    verbose=VERBOSE,
)
```
