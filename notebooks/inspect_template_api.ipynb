# Microscopy Template Inspection — API Workflow

Connects to a live microscope instance, then parses, enriches, and
visualises the scanning template.

**Requirements:** microscope software running with the API server enabled.

Uses the manufacturer-agnostic `initialize_api()` and
`initialize_experiment()` entry points. The `BACKEND` parameter selects
the microscope vendor (currently `"lasx"` for Leica LAS X).

```python
# =============================================================================
# API Connection
# =============================================================================
#
# Establishes a connection to the microscope API.
# The connector object is reused in the inspection cell below.
#
# BACKEND     : Microscope vendor backend.
# CLIENT_NAME : Identifier for this connection (shown in the software).
# TIMEOUT     : Seconds to wait for API responses.
# =============================================================================

import sys
from pathlib import Path

sys.path.insert(0, str((Path("..") / "lib").resolve()))

from microscope_connector import initialize_api

BACKEND     = "lasx"
CLIENT_NAME = "PythonClient"
TIMEOUT     = 15.0

connector = initialize_api(BACKEND, client_name=CLIENT_NAME, timeout=TIMEOUT)

if connector.is_connected:
    print(f"Connected to {BACKEND!r} API.")
    print(f"Ready -- the connector is passed to the inspection cell below.")
else:
    print("Connection failed. Check that the microscope software is running")
    print("with the API server enabled.")
```

```python
%matplotlib inline
from initialize_experiment import initialize_experiment

# -- Parameters ---------------------------------------------------------------
INPUT           = "auto"        # "auto" | path to template folder | None for picker
OUTPUT_PNG      = None          # save figure + JSON, e.g. "./out.png"
FIGSIZE         = (14, 10)      # figure size (width, height) in inches
VERBOSE         = 1             # 0 = data only | 1 = summary + figure | 2 = full report

# -- Run ---------------------------------------------------------------------
data = initialize_experiment(
    BACKEND,
    INPUT,
    enrich="api",
    client=connector,
    output_path=OUTPUT_PNG,
    figsize=FIGSIZE,
    verbose=VERBOSE,
)
```
